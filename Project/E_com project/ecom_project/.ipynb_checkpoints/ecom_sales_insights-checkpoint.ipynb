### Step-2 ) Data Cleaning

```python
df['PaymentMode'] = df['PaymentMode'].str.lower().replace({
    "cash": "Cash", 
    "card": "Card", 
    "credit card": "Card", 
    "upi": "UPI",
    "online": "Online"
})
```

```python
df
```

```python
# Missing values
df.isnull().sum()
```

```python
df['Discount'] = df['Discount'].fillna(0)
```

```python
df
```

```python
df['City'] = df['City'].fillna("Unknown")
df
```

```python
df = df.dropna(subset=['CustomerName'])
df
```

### Step 3: GroupBy & Aggregation

#### 1) Total Sales Value per City

```python
df['NetPrice'] = df['Price'] * df['Quantity'] * (1 - df['Discount']/100)
sales_per_city = df.groupby('City')['NetPrice'].sum()
```

```python
print(sales_per_city)
```

#### 2) Category Stats (Total Qty, Avg Price, Max Discount)

```python
category_stats = df.groupby('Category').agg(
    Total_Qty=('Quantity','sum'),
    Avg_Price=('Price','mean'),
    Max_Discount=('Discount','max')
)
print(category_stats)
```

#### 3) Top 3 Customers (by spend)

```python
top_customers = df.groupby('CustomerName')['NetPrice'].sum().nlargest(3)
```

```python
print(top_customers)
```

#### 4) Most Popular Product per City

```python
popular_products = df.groupby(['City','Product'])['Quantity'].sum().reset_index()
popular_products = popular_products.loc[popular_products.groupby('City')['Quantity'].idxmax()]
print(popular_products)
```

## Step 4: Advanced Analysis (Part C)

### 1) Revenue Column & Monthly Trend

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
df['OrderDate'].dtype
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')
df
```

```python
df['OrderDate'].unique()
```

```python
df['Revenue'] = df['Quantity'] * df['Price'] * (1 - df['Discount']/100)
df['OrderDate'] = pd.to_datetime(df['OrderDate'])  # ensure datetime
monthly_revenue = df.groupby(df['OrderDate'].dt.to_period('M'))['Revenue'].sum()
monthly_revenue.plot(kind='line', title='Monthly Revenue Trend')
```

### 2) Average Revenue per Order by Payment Mode

```python
revenue_by_payment = df.groupby('PaymentMode')['Revenue'].mean()
print(revenue_by_payment)
```

### 3) Category-wise Revenue Share per City

```python
city_category_share = df.groupby(['City','Category'])['Revenue'].sum().reset_index()
city_totals = df.groupby('City')['Revenue'].sum().reset_index()
city_category_share = city_category_share.merge(city_totals, on='City', suffixes=('', '_Total'))
city_category_share['Share%'] = (city_category_share['Revenue'] / city_category_share['Revenue_Total']) * 100
```

### 4) Customers Buying from â‰¥3 Categories

```python
multi_category_customers = df.groupby('CustomerName')['Category'].nunique()
multi_category_customers = multi_category_customers[multi_category_customers >= 3]
```

## Step 5: Visualization & Dashboard

### 1. Monthly Revenue Trend

```python
import matplotlib.pyplot as plt

monthly_revenue.index = monthly_revenue.index.to_timestamp()
monthly_revenue.plot(kind='line', marker='o', figsize=(10,5), title='Monthly Revenue Trend')
plt.ylabel("Revenue")
plt.xlabel("Month")
plt.show()
```

### 2. Revenue Share by Payment Mode

```python
payment_share = df.groupby('PaymentMode')['Revenue'].sum()
payment_share.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6), title="Revenue Share by Payment Mode")
plt.ylabel("")
plt.show()
```

### 3. Top 5 Customers by Spend

```python
top_customers = df.groupby('CustomerName')['Revenue'].sum().nlargest(5)
top_customers.plot(kind='bar', figsize=(8,5), title="Top 5 Customers by Revenue")
plt.ylabel("Revenue")
plt.show()
```

### 4. Category-wise Revenue per City (Stacked Bar)

```python
city_category = df.pivot_table(values='Revenue', index='City', columns='Category', aggfunc='sum', fill_value=0)
city_category.plot(kind='bar', stacked=True, figsize=(10,6), title="Category-wise Revenue per City")
plt.ylabel("Revenue")
plt.show()
```
