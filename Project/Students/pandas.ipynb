# 5: Data Manipulation 

```python
# Add Column
df['Total'] = df[['Math', 'Science', 'English']].sum(axis=1)
df
```

```python
# Delete column
df.drop('Total', axis=1, inplace=True)
df
```

```python
# Apply Function using apply()
def get_grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 75:
        return 'B'
    elif avg >= 60:
        return 'C'
    else:
        return 'D'

df['Grade'] = df[['Math', 'Science', 'English']].mean(axis=1).apply(get_grade)
```

```python
#Use map() with Lambda
df['Gender'] = df['Gender'].map({'Male': 'M', 'Female': 'F'})
```

```python
#   Use lambda with apply()
df['Passed'] = df['Math'].apply(lambda x: 'Yes' if x >= 35 else 'No')
```

# GroupBy and Aggregation in Pandas

```python
import pandas as pd

data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi'],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female', 'Male', 'Female', 'Female'],
    'Class': ['10A', '10B', '10A', '10B', '10A', '10B', '10A', '10B'],
    'Math': [85, 78, 92, 67, 89, 72, 90, 76],
    'Science': [88, 75, 85, 70, 90, 80, 84, 78]
}

df = pd.DataFrame(data)
df
```

```python
# Use groupby() to Group Data
grouped = df.groupby('Class')
```

```python
#  Apply Aggregation Functions
grouped[['Math', 'Science']].sum()
grouped[['Math', 'Science']].mean() # Mean Marks per Class
grouped['Student'].count()  #Count of Students per Class
```

```python
# Group by Multiple Columns
df.groupby(['Class', 'Gender'])[['Math', 'Science']].mean()
```

```python
# Use .agg() for Multiple Aggregations
df.groupby('Class').agg({
    'Math': ['mean', 'max'],
    'Science': 'sum'
})
```

# 7: Merging, Joining, and Concatenating in Pandas

```python
import pandas as pd

# Student basic info
df_students = pd.DataFrame({
    'StudentID': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Class': ['10A', '10A', '10B', '10B']
})

# Student scores
df_scores = pd.DataFrame({
    'StudentID': [101, 102, 104, 105],
    'Math': [85, 90, 78, 88],
    'Science': [89, 92, 84, 75]
})
```

```python
df_students
```

```python
df_scores
```

```python
# merge() – Combining on Key Column
pd.merge(df_students, df_scores, on='StudentID')
```

```python
# Outer Join
pd.merge(df_students, df_scores, on='StudentID', how='outer')
pd.merge(df_students, df_scores, on='StudentID', how='left') # Left join
pd.merge(df_students, df_scores, on='StudentID', how='right')  # Right Join
  
```

```python
# concat() – Stack DataFrames
# Vertical (Row-wise) Concatenation

df1 = pd.DataFrame({'ID': [1, 2], 'Name': ['A', 'B']})
df2 = pd.DataFrame({'ID': [3, 4], 'Name': ['C', 'D']})

pd.concat([df1, df2], ignore_index=True)
```

```python
# Horizontal (Column-wise) Concatenation
df1 = pd.DataFrame({'ID': [1, 2]})
df2 = pd.DataFrame({'Marks': [80, 90]})

pd.concat([df1, df2], axis=1)
```

```python
# join() – Using Index
df1 = df_students.set_index('StudentID')
df2 = df_scores.set_index('StudentID')

df1.join(df2, how='inner')
```
